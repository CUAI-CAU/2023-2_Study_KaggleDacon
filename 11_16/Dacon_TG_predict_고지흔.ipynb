```python
def interpolate(df, flag):
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    # 모든 날짜를 포함하는 새로운 날짜 범위를 생성
    full_date_range = pd.date_range(start=df.index.min(), end=df.index.max())
    # 새로운 날짜 범위를 사용하여 데이터 프레임을 재인덱싱
    df = df.reindex(full_date_range)
    # 선형 보간을 사용하여 빈 데이터를 채움
    df = df.interpolate(method='linear')
    df = df.reset_index().rename(columns={'index': 'timestamp'})
    if flag:
        df['twentyfour']=df['twentyfour']%24
    df.rename(columns={'Date': 'timestamp'}, inplace=True)
    return df
```

```python
def add_time_feature(df):
    df['timestamp'] = pd.to_datetime(df['timestamp'])
#     df['month_sin']=np.sin((df['month']-1)/12)
#     df['month_cos']=np.cos((df['month']-1)/12)

    df['year'] = df['timestamp'].dt.year
    df['month'] = df['timestamp'].dt.month
    df['day'] = df['timestamp'].dt.day

    # 요일 추가 (0: 월요일, ..., 6: 일요일)
    df['weekday'] = df['timestamp'].dt.dayofweek
    df['weekday_name'] = df['timestamp'].dt.day_name().astype('category')  # 요일 이름
    

    # 주말 여부 추가 (토요일과 일요일은 주말로 간주)
    df['is_weekend'] = df['weekday'].apply(lambda x: True if x >= 6 else False)
    df['is_weekend'] = df['is_weekend'] | df['timestamp'].isin(Holiday)
    df=df.drop('weekday',axis=1)

    # 1년 중 몇 번째 날 추가
    df['day_of_year'] = df['timestamp'].dt.dayofyear
#     df['day_of_year_sin']=np.sin((df['day_of_year']-1)/365)
#     df['day_of_year_cos']=np.cos((df['day_of_year']-1)/365)

    # 1년 중 몇 번째 주 추가 (ISO 표준에 따름)
    df['week_of_year'] = df['timestamp'].dt.isocalendar().week
    df['week_of_year'] = df['week_of_year'].astype(int)
#     df = pd.merge(df, solar_terms_df, on='timestamp', how='left')
    df = pd.merge(df, temp_solar_terms_df, on='timestamp', how='left')
 
    return df
```

```python
# 월별 평균 가격 계산
monthly_avg = train.groupby(['item', 'year', 'month'])['price'].mean().reset_index()

# 2019년 데이터에 해당 연도의 월별 평균 가격 추가
train['price_prev_year_avg'] = None  # 새 컬럼을 초기화합니다.

for year in range(2019, 2024):
    for month in range(1, 13):
        # 2023년 4월 이후 데이터는 불필요
        if year == 2023 and month > 3:
            break
        if year == 2019:
            # 2019년에는 해당 연도의 평균을 사용
            price_avg = monthly_avg[(monthly_avg['year'] == year) & (monthly_avg['month'] == month)]
        else:
            # 2020년 이후에는 이전 연도의 같은 월 평균을 사용
            price_avg = monthly_avg[(monthly_avg['year'] == year - 1) & (monthly_avg['month'] == month)]
        price_avg_dict = price_avg.set_index('item')['price'].to_dict()
        mask = (train['year'] == year) & (train['month'] == month)
        train.loc[mask, 'price_prev_year_avg'] = train.loc[mask, 'item'].map(price_avg_dict)
        
# test 데이터에 2022년 3월의 품목별 평균 가격 추가
march_2022_avg = monthly_avg[(monthly_avg['year'] == 2022) & (monthly_avg['month'] == 3)][['item', 'price']].rename(columns={'price': 'price_prev_year_avg'})
test = test.merge(march_2022_avg, on='item', how='left')
```

```python
# 월별 평균 가격 계산
monthly_avg = train.groupby(['item', 'year', 'month'])['price'].mean().reset_index()

# train 데이터셋에 품목별 이전 월의 평균 가격 추가
train['price_prev_month_avg'] = None  # 새 컬럼을 초기화합니다.

# 2019년 1월 데이터에 대해 처리
jan_2019_avg = monthly_avg[(monthly_avg['year'] == 2019) & (monthly_avg['month'] == 1)]
jan_2019_avg_dict = jan_2019_avg.set_index('item')['price'].to_dict()
train.loc[(train['year'] == 2019) & (train['month'] == 1), 'price_prev_month_avg'] = train['item'].map(jan_2019_avg_dict)

# 나머지 데이터에 대한 이전 월 평균 가격 계산
for year in range(2019, 2024):
    for month in range(1, 13):
        # 2023년 4월 데이터는 필요 없으므로 반복문을 끝냄
        if year == 2023 and month > 3:
            break
        # 2019년 1월은 이미 처리했으므로 스킵
        if year == 2019 and month == 1:
            continue
        
        # 1월의 경우, 이전 해의 12월 평균 사용
        prev_year = year if month > 1 else year - 1
        prev_month = month - 1 if month > 1 else 12
        
        # 이전 월 평균 가격 추출
        prev_month_avg = monthly_avg[
            (monthly_avg['year'] == prev_year) & 
            (monthly_avg['month'] == prev_month)
        ]
        prev_month_avg_dict = prev_month_avg.set_index('item')['price'].to_dict()

        # 해당 월 데이터에 이전 월 평균 가격 매핑
        mask = (train['year'] == year) & (train['month'] == month)
        train.loc[mask, 'price_prev_month_avg'] = train.loc[mask, 'item'].map(prev_month_avg_dict)

# test 데이터에 2023년 2월의 품목별 이전 월 평균 가격 추가
feb_2023_avg = monthly_avg[(monthly_avg['year'] == 2023) & (monthly_avg['month'] == 2)]
feb_2023_avg_dict = feb_2023_avg.set_index('item')['price'].to_dict()
test['price_prev_month_avg'] = test['item'].map(feb_2023_avg_dict)
```
